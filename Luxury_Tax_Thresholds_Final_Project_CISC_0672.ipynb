# **Displays the Luxury Tax Thresholds Between 2012 - 2025**
Major League Baseball’s Competitive Balance Tax (CBT), commonly called the luxury tax, is designed to limit excessive payroll spending. Each season, MLB sets a payroll threshold. Teams whose CBT payroll exceeds that amount must pay escalating tax rates depending on how many consecutive years they have gone over the limit.

**Base Tax Rates**

First year over: 20%

Second consecutive year: 30%

Third+ consecutive years: 50%

*IMPORANT NOTE: Going under the threshold for one season resets a team to the 20% rate.*


**Surcharge Tiers**

Additional penalties apply based on how far a team exceeds the threshold:

$0–20M over: base rate only

$20–40M over: +12% surcharge (total becomes 32%, 42%, or 62%)

$40–60M over: +42.5% surcharge for first-time offenders, +45% for repeat offenders

$60M+ over (“Steve Cohen Tax”): +60% surcharge and a 10-spot draft pick drop the following year.

Steve Cohen is a billionaire hedge fund manager and owner of the New York Mets. The Mets are a team notorious for spending well beyond the luxury tax threshold.

# **Creates Total Spending Trends Graph (CBT Payroll + Luxury Tax)**

**MLB Payroll** is the total amount of money a baseball team spends on its players. But when deciding if a team has to pay the luxury tax, MLB does not look at the exact dollar amount paid that year.

Instead, it uses a special number called the **“CBT payroll.”**

The CBT payroll is based on the average value of each player’s contract, not the year-to-year salary.

**Example:** A 5-year, 100 million contract counts as 20 million every year, even if the player gets paid more or less in certain seasons.

**MLB adds up:**

The average salary of all players on the 40-man roster

Player benefits (insurance, pensions, etc.)

Certain minor-league salaries

Any money the team is still paying to traded players

This total is the team’s CBT payroll.



**When Luxury Tax is added to the CBT Payroll this is a Team's Total Spending**

```python
import altair as alt
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# ---- TEAM SELECTOR WIDGET ----
team_selector = widgets.SelectMultiple(
    options=sorted(df_payroll['Team'].unique()),
    description="Teams",
    layout=widgets.Layout(width='300px', height='500px')
)

# ---- YEAR RANGE SLIDER (2012–2025) ----
year_slider = widgets.IntRangeSlider(
    value=[2012, 2025],
    min=2012,
    max=2025,
    step=1,
    description='Years:',
    continuous_update=False,
    layout=widgets.Layout(width='300px')
)

# ---- OUTPUT AREA ----
plot_out = widgets.Output()

# ---- UPDATE FUNCTION ----
def update_plot(change):
    with plot_out:
        clear_output(wait=True)

        selected = list(team_selector.value)
        start_year, end_year = year_slider.value

        if not selected:
            print("Select teams to display. Hold CTRL to select multiple teams.")
            return

        # Apply BOTH filters
        df_filtered = df_payroll[
            (df_payroll['Team'].isin(selected)) &
            (df_payroll['Year'].between(start_year, end_year))
        ]

        chart = (
            alt.Chart(df_filtered)
            .mark_line(
                point=alt.OverlayMarkDef(
                    size=150,
                    filled=True,
                    shape='square'
                )
            )
            .encode(
                x=alt.X(
                    'Year:O',
                    title='Year',
                    axis=alt.Axis(
                        titleFontSize=22,   # X-axis title font size
                        labelFontSize=14
                    )
                ),
                y=alt.Y(
                    'Tax Payroll:Q',
                    title='Total Spending ($)',
                    axis=alt.Axis(
                        titleFontSize=22,   # Y-axis title font size
                        labelFontSize=14
                    )
                ),
                color='Team:N',
                tooltip=[
                    alt.Tooltip('Team:N', title='Team'),
                    alt.Tooltip('Year:O', title='Year'),
                    alt.Tooltip('Tax Payroll:Q', title='Total', format=',')
                ]
            )
            .properties(
                width=1200,
                height=750,
                title=alt.TitleParams(
                    "Total Spending Each Year (CBT Payroll + Luxury Tax) (2012–2025)",
                    fontSize=32,        # MAIN TITLE FONT SIZE (edit here!)
                    fontWeight="bold"
                )
            )
        )

        display(chart)

# React to widget changes
team_selector.observe(update_plot, names='value')
year_slider.observe(update_plot, names='value')

# ---- UI LAYOUT ----
controls = widgets.VBox([team_selector, year_slider])

ui = widgets.HBox([
    plot_out,
    controls
])

display(ui)

# Load initial empty view
update_plot(None)

```

```python
import altair as alt
import pandas as pd

# -----------------------------------
# CLEAN CAP SPACE COLUMN
# -----------------------------------
df_payroll['Cap Space'] = (
    df_payroll['Cap Space']
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.replace('(', '-', regex=False)
        .str.replace(')', '', regex=False)
)

df_payroll['Cap Space'] = pd.to_numeric(df_payroll['Cap Space'], errors='coerce').fillna(0)
df_payroll['Year'] = df_payroll['Year'].astype(int)

# -----------------------------------
# COMPUTE "Exceeded Threshold" (positive means above the threshold)
# -----------------------------------
df_payroll['Exceeded Amount'] = df_payroll['Cap Space'].apply(lambda x: max(0, -x))

# -----------------------------------
# YEAR SLIDER
# -----------------------------------
year_slider = alt.binding_range(
    min=df_payroll['Year'].min(),
    max=df_payroll['Year'].max(),
    step=1,
    name='Select Year: '
)

year_select = alt.selection_point(
    fields=['Year'],
    bind=year_slider
)

# -----------------------------------
# BAR CHART — ALL TEAMS
# -----------------------------------
chart = (
    alt.Chart(df_payroll)
    .mark_bar()
    .encode(
        x=alt.X('Team:N', title='MLB Teams', sort='-y'),
        y=alt.Y('Cap Space:Q', title='Cap Space (USD; Negative = Over Threshold)'),
        color=alt.condition(
            alt.datum['Cap Space'] < 0,
            alt.value('red'),    # over threshold
            alt.value('green')   # under threshold
        ),
        tooltip=[
            alt.Tooltip('Team:N', title='Team'),
            alt.Tooltip('Year:O', title='Year'),
            alt.Tooltip('Cap Space:Q', title='Cap Space', format='$,.0f'),
            alt.Tooltip('Exceeded Amount:Q', title='Amount Over Threshold', format='$,.0f')
        ]
    )
    .add_params(year_select)
    .transform_filter(year_select)
    .properties(
        width=1200,
        height=750,
        title='MLB Team Cap Space Relative to Spending Threshold'
    )
)

chart



```

# **Description of the Graphs (2012–2025)**

These graphs show MLB team cap space relative to the luxury tax threshold for every season from 2012 through 2025, highlighting how spending behavior has evolved over time.

In the early years—especially around 2012—nearly all teams operated comfortably below the tax threshold, with only one or two exceptions. Most clubs maintained large positive cap space and followed relatively conservative payroll strategies.

As the years progress toward 2025, the trend shifts sharply. More teams begin to creep closer to the threshold, several start to hover around zero, and by the later seasons, a growing number of franchises are willing to overspend—sometimes massively. By 2025, high-spending organizations like the Dodgers, Mets, Yankees, Phillies, Blue Jays, Padres, Red Sox, and Astros operate far over the limit, demonstrating an aggressive willingness to pay substantial luxury tax penalties.

MLB has moved from widespread financial restraint to an era where many teams aggressively exceed the luxury tax threshold in pursuit of competitive advantage.

```python
import altair as alt
import pandas as pd

# --- CLEAN "Tax Payroll" COLUMN AND YEAR ---
df_payroll['Tax Payroll'] = (
    df_payroll['Tax Payroll']
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.replace('(', '-', regex=False)
        .str.replace(')', '', regex=False)
)

df_payroll['Tax Payroll'] = pd.to_numeric(df_payroll['Tax Payroll'], errors='coerce')
df_payroll['Year'] = df_payroll['Year'].astype(int)

# --- FILTER ONLY 2012 & 2025 ---
df_2yr = df_payroll[df_payroll['Year'].isin([2012, 2025])].copy()

# Aggregate in case duplicates exist
team_totals = (
    df_2yr
    .groupby(['Team', 'Year'], as_index=False)['Tax Payroll']
    .sum()
)

# --- PIVOT TO GET 2012 & 2025 SIDE BY SIDE ---
pivot = (
    team_totals
    .pivot(index='Team', columns='Year', values='Tax Payroll')
    .dropna(subset=[2012, 2025])   # keep only teams with both years
)

# --- PERCENT CHANGE: (2025 - 2012) / 2012 * 100 ---
pivot['Pct_Change'] = (pivot[2025] - pivot[2012]) / pivot[2012] * 100

df_pct = pivot.reset_index()[['Team', 'Pct_Change']]

# --- BAR CHART (GREEN ABOVE, RED BELOW X-AXIS) ---
chart = (
    alt.Chart(df_pct)
    .mark_bar()
    .encode(
        x=alt.X(
            'Team:N',
            sort='-y',
            title='MLB Teams',
            axis=alt.Axis(
                titleFontSize=26   # X-axis title font size
            )
        ),
        y=alt.Y(
            'Pct_Change:Q',
            title='Percent Change in Total Spending (2012 → 2025) (%)',
            axis=alt.Axis(
                format='.1f',
                titleFontSize=26   # Y-axis title font size
            )
        ),
        color=alt.condition(
            alt.datum.Pct_Change > 0,
            alt.value('green'),   # positive = green
            alt.value('red')      # negative = red
        ),
        tooltip=[
            alt.Tooltip('Team:N', title='Team'),
            alt.Tooltip('Pct_Change:Q', title='% Change (2012→2025)', format='.2f')
        ]
    )
    .properties(
        width=1100,
        height=800,
        title=alt.TitleParams(
            'MLB Total Spending Percentage Change: 2012 vs 2025',
            fontSize=30,     # <<-- make this bigger/smaller as you like
            fontWeight='bold'
        )
    )
)

chart
```

```python
import altair as alt
import pandas as pd

# Prepare Champion data
df_champions = df_WS.copy()
df_champions = df_champions.rename(columns={'WS Champion': 'Team'})
df_champions['Result'] = 'Champion'

# Prepare Runner-up data
df_runnerups = df_WS.copy()
df_runnerups = df_runnerups.rename(columns={'WS Runner Up': 'Team'})
df_runnerups['Result'] = 'Runner-up'

# Combine champion and runner-up data
df_ws_teams = pd.concat([df_champions[['Year', 'Team', 'Result']], df_runnerups[['Year', 'Team', 'Result']]])

# Merge with payroll data to get 'Tax Payroll'
df_merged = pd.merge(df_ws_teams, df_payroll[['Year', 'Team', 'Tax Payroll']], on=['Year', 'Team'], how='left')

# Clean 'Tax Payroll' column if necessary (already done in previous cells, but good to be explicit)
df_merged['Tax Payroll'] = (
    df_merged['Tax Payroll']
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.replace('(', '-', regex=False)
        .str.replace(')', '', regex=False)
)
df_merged['Tax Payroll'] = pd.to_numeric(df_merged['Tax Payroll'], errors='coerce')

# Create the double bar chart
chart = (
    alt.Chart(df_merged)
    .mark_bar()
    .encode(
        x=alt.X(
            'Year:O',
            axis=alt.Axis(
                title='Year',
                labelAngle=0
            )
        ),
        xOffset='Result:N',
        y=alt.Y(
            'Tax Payroll:Q',
            title='Spending ($)',
            axis=alt.Axis(
                format='$,.0f'
            )
        ),
        color=alt.Color(
            'Result:N',
            scale=alt.Scale(domain=['Champion', 'Runner-up'], range=['blue', 'orange']),
            title='World Series Result'
        ),
        tooltip=[
            alt.Tooltip('Team:N', title='Team'),
            alt.Tooltip('Year:O', title='Year'),
            alt.Tooltip('Tax Payroll:Q', title='Spending', format='$,.0f')
        ]
    )
    .properties(
        width=800,
        height=500,
        title=alt.TitleParams(
            'WS Champion vs Runner-Up Spending by Year',
            fontSize=20,
            anchor='start'
        )
    )
)

chart
```
